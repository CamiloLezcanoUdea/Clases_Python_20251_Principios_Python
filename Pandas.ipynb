```python
# pandas
import pandas as pd
import numpy as np
```

```python
# 1. crear dataFrame

df1 = pd.DataFrame( {'Nombre': ['Ana','Luis','Fernando'], 'Edad': [23, 25, 28] })
print(df1)
```

```python
# 2. crear dataFrame a partir de una lista

df2 = pd.DataFrame([['Ana',23], ['Luis', 25], ['Fernando', 28]], columns=['Nombre', 'Edad'])
print(df2)
```

```python
# 3. crear dataFrame con indices personalizados

df3 = pd.DataFrame({'Nombre': ['Ana', 'Luis', 'Fernando']}, index=['a', 'b', 'c'])
print(df3)
```

```python
# 4. ver las primeras filas

df1.head()
```

```python
# 5. ver los ultimos valores

df1.tail()
```

```python
# 6. informacion general

df1.info()
```

```python
# 7. ver tipos de datos

df1.dtypes
```

```python
# 8. cambiar nombre de columnas

df1.columns = ['Nombre_Completo', 'Edad_Relativa']
print(df1)
```

```python
# 9. Selección de una columna

names = df1['Nombre_Completo']
print(names)
```

```python
# 10. Selección de multiples columnas

df1[['Nombre_Completo', 'Edad_Relativa']]
```

```python
# 11. Filtrar por condición

df1[df1['Edad_Relativa'] > 23]
df1[df1['Nombre_Completo'] != "Ana"]
```

```python
# 12. Usar loc

df1.loc[0]
df1.iloc[0]
```

```python
# 13. filtrar con multiples condicionales

operatorOr = df1[(df1['Edad_Relativa'] < 28) | (df1['Nombre_Completo'] != 'Ana')]
operatorAnd = df1[(df1['Edad_Relativa'] < 28) & (df1['Nombre_Completo'] != 'Ana')]

print(operatorOr)
print(operatorAnd)
```

```python
# 14. Asignar nuevos valores a las columnas

df1['Ciudad'] = ['Madrid', 'Barcelona', 'Valencia']
print(df1)
```

```python
# 15. Calcular la media

df1['Edad_Relativa'].mean()
```

```python
# 16. Calcular la mediana

df1['Edad_Relativa'].median()
```

```python
# 17. Calcular la desviación estandar

df1['Edad_Relativa'].std()
```

```python
# 18. Calcular el valor máximo

df1['Edad_Relativa'].max()
```

```python
# 19. Calcular el valor mínimo

df1['Edad_Relativa'].min()
```

```python
# 20. Descripción estadística

df1.describe()
```

```python
# 21. Reemplazar valores

df1['Ciudad'] = df1['Ciudad'].replace(['Madrid', 'Barcelona', 'Valencia'], ['Bogota', 'Cali', 'Medellin'])
print(df1)
```

```python
# 22. Eliminar una columna

df1 = df1.drop(columns='Ciudad')
print(df1)
```

```python
# 23. Eliminar una fila

df1 = df1.drop(index=1)
print(df1)
```
